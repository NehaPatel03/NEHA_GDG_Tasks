#Bar charts

```python
# LOGIC :
# 1. First, we'll use groupby() method to group the data based on the Sex column.
# 2. Select the Age column from the groups of Sex column( M, F, I) created by the groupby() method.
# 3. Calculate the average age for each sex using the mean() method.
```

```python
# plot the graph here
average_age_by_sex = crabs_data.groupby('Sex')['Age'].mean()
plt.bar(average_age_by_sex.index, average_age_by_sex.values)
plt.xlabel('Sex')
plt.ylabel('Average Age')
plt.title('Average Age by Sex')
plt.show()

```

#Box plots

```python
# plot the graph here
plt.figure(figsize=(8, 6))
sns.boxplot(x='Sex', y='Age', data=crabs_data)
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Boxplot of Age by Sex')
plt.show()
```

#KDE plots

```python
# write code that yields output similar to the one shown below
plt.figure(figsize=(16.38, 8.4))
# For Age
plt.subplot(1, 3, 1)  # 1 row, 3 columns, first subplot
sns.kdeplot(crabs_data['Age'], fill=True, color='blue')
plt.xlabel('Age')
plt.ylabel('Density')

# For Length
plt.subplot(1, 3, 2)  # 1 row, 3 columns, second subplot
sns.kdeplot(crabs_data['Length'], fill=True, color='orange')
plt.xlabel('Length')
plt.ylabel('Density')


# For Weight
plt.subplot(1, 3, 3)  # 1 row, 3 columns, third subplot
sns.kdeplot(crabs_data['Weight'], fill=True, color='green')
plt.xlabel('Weight')
plt.ylabel('Density')


plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# write code here
crabs_data.insert(3, 'Lost Weight', crabs_data['Weight'] - (crabs_data['Shucked Weight'] + crabs_data['Viscera Weight'] + crabs_data['Shell Weight']))
crabs_data.loc[crabs_data['Lost Weight'] <= 0, 'Lost Weight'] = 0
crabs_data.loc[crabs_data['Lost Weight'] > 0, 'Lost Weight'] = 1
crabs_data.head()
```

```python
# Joint Point
joint_plot = sns.jointplot(
  data=crabs_data,
  x='Length',
  y='Weight',
  hue='Sex',
  palette='bright',
  marginal_kws=dict(fill=True),
)
joint_plot.fig.suptitle("Joint Plot of Crab Length and Weight", y=1.02, fontsize=16)
plt.show()
```

Machine Learning algorithms can only crunch numbers. Give it a string or text input, and it self-annihilates by throwing an error. To handle this, we use a technique called one hot encoding.

```python
# write code here
crabs_data = pd.get_dummies(crabs_data, columns=['Sex'])
crabs_data.head()
```

#Normalisation and standardisation

To prevent some data features from dominating the model training process, the above two feature scaling techniques is to be implemented.

Normalise the dataset.

```python
# write code here
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_norm = pd.DataFrame(scaler.fit_transform(crabs_data), columns=crabs_data.columns)
df_norm.head()

```

For standardisation, the one hot encoded columns are not standardised. They are categorical in nature, so it makes no sense to shoehorn them into any type of distribution at all.

It will be scaled down to values between 0 and 1, which is the range in which they already exist anyways. Normalisation doesn't affect them.

```python
# since crabs_data is already one hot encoded for the column 'Sex'
crabs_data.head()
```
